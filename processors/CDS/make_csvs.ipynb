```python
import numpy as np
import pandas as pd
```

```python
import xarray as xr
```

```python
var_name = 'PRECIP'
```

```python
dict_vars = {}
dict_vars['PRECIP'] = 'tprate'
dict_vars['SST'] = 'sst'
dict_vars['T2M'] = 't2m'
dict_vars['Z850'] = 'z'
```

```python
dset = xr.open_dataset(f'/home/nicolasf/drives/auck_projects/END19101/Working/data/hindcasts/ecmwf/{var_name}/{var_name}_hindcasts_1993_2016_ecmwf_seasonal_anomalies.nc')
```

```python
dsetm = dset.mean('number')
```

```python
dsetm = dsetm.sortby('latitude')
```

```python
dsetm = dsetm.sel(latitude=slice(None, 40))
```

```python
one_month_lead = dsetm.sel(step=3).squeeze()
```

```python
data_arr = one_month_lead[dict_vars[var_name]]
```

```python
df = data_arr.to_dataframe()
```

```python
df = df.loc[:,[dict_vars[var_name]]]
```

```python
df_unstacked = df.unstack(level=['longitude','latitude'])
```

```python
df_unstacked.shape
```

```python
df_unstacked.head()
```

```python
time = df_unstacked.index
```

```python
from dateutil.relativedelta import relativedelta
```

```python
time = [x + relativedelta(months=3) for x in time]
```

```python
df_unstacked.index = time
```

```python
df_unstacked.head()
```

```python
df_unstacked = df_unstacked.dropna(axis=1)
```

```python
df_unstacked.to_csv(f'/home/nicolasf/drives/auck_projects/END19101/Working/outputs/ECMWF_CSVs/{var_name}_1993_2016_ecmwf.csv') 
```
